```python
#課題内容：8.4章のモデルを決定木→ランダムフォレストに変更
#1.load_wineのデータセットを取得して、dataset変数に入れる、データの中身を確認
#2.データセットを表形式にする（このときpandasのDataFrameメソッドを使用する）
#3.datasetのdataとtargetをそれぞれx,yにいれ、学習用とtrain用に分ける
#4.学習モデルの種類を決める（今回はランダムフォレスト）,モデルをfitさせる
#5.予測モデルの評価をする
```

```python
#1.load_wineのデータセットを取得して、dataset変数に入れる、データの中身を確認
from sklearn.datasets import load_wine
```

```python
dataset=load_wine()
```

```python
#2.データセットを表形式にする（このときpandasのDataFrameメソッドを使用する）
import pandas as pd
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df['category']=dataset.target
df.head()
```

```python
#3.datasetのdataとtargetをそれぞれx,yにいれ、学習用とtrain用に分ける

#x=pd.DataFrame(dataset.data)
#y=pd.DataFrame(dataset.target)
X=df[['alcohol','malic_acid','ash','alcalinity_of_ash','magnesium','total_phenols','flavanoids','nonflavanoid_phenols','proanthocyanins','color_intensity','hue','od280/od315_of_diluted_wines','proline']]
y=df['category']
```

```python
from sklearn.model_selection import train_test_split
train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
print(X.shape, X_train.shape, X_test.shape, y.shape, y_train.shape, y_test.shape)
```

```python
#4.学習モデルの種類を決める（今回はランダムフォレスト）,モデルをfitさせる
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(random_state=3)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
#5.予測モデルの評価をする(今回は正解率 accuracy)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
