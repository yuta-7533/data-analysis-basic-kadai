```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

df1=pd.read_csv('sample_pandas_6.csv')
df2=pd.read_csv('category.csv')
print(df1.head())
print(df1.shape)
print(df2)
print(df2.shape)
```

```python
df=pd.merge(df1,df2[['商品番号','カテゴリー']],how='inner',on='商品番号')
df.head()
```

```python
#カテゴリー列の要素(弁当、デザートなど)の出現頻度をカウント
counts=df['カテゴリー'].value_counts()
print(counts)

#1.出現頻度の結果を棒グラフで表示する
counts.plot(kind='bar')

plt.show()
```

```python
#商品番号列の商品番号ごとに、注文数の基本統計量を求める
statistics=df.groupby('商品番号')['注文数'].describe()
print(statistics)
```
