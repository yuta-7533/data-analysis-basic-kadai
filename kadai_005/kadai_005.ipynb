```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2,3,tight_layout=True)

#折れ線グラフの描画
x=[0,1,2,3]
y=[1,2,-5,2]
ax[0,0].plot(x,y)

#sinのグラフ描画
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# Select the first axes object in the array to plot on
ax[0, 1].plot(x, y, linewidth=2.0)

ax[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
           ylim=(-1, 5), yticks=np.arange(-1, 6))

#ヒストグラムの描画
# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x = np.random .normal(15, 5, 2000)
# ヒストグラムをプロット
ax[0,2].hist(x, bins=10)

#散布図の描画
# データを生成する
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

# プロットする
ax[1,0].scatter(x, y)


plt.show()
```
